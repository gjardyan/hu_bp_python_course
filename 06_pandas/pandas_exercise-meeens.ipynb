### Task 5: Plot the age composition of the different BundeslÃ¤nder

```python
ageperb = bundeslaender.groupby('NAME')['ALTER_1','ALTER_2','ALTER_3','ALTER_4','ALTER_5'].sum()
ageperb.plot(kind='barh', stacked=True)
```

### Task 6: ...and now relative to the population in each Bundesland

```python
bnd = bundeslaender.set_index('NAME')
rel = ageperb.div(bnd['EWZ'], axis = 0)
rel.plot(kind='barh', stacked=True)
```

### Task 7: Plot pie charts of the relative age distribution in Berlin and Sachsen

```python
berlin = rel.loc['Berlin']
sachsen = rel.loc['Sachsen']

berlin.plot(kind = 'pie')
sachsen.plot(kind = 'pie')
```
