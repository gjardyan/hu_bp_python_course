```python
import pandas as pd
from matplotlib import pyplot as plt
plt.style.use('ggplot')
%matplotlib inline
diamonds = pd.read_csv('diamonds.csv',index_col=0)
diamonds
```

### Task 1: Find all the diamonds with more than 2 carat and plot their price distribution in a histogram

```python
diamondscarat = diamonds[diamonds['carat']>2]
ddp = diamondscarat.dropna( axis=0, how='any')
ddp["price"].plot(kind="hist")
```

### Task 2: compute and plot the standard deviation of the x dimension for the different cuts

```python
ddd = ddp.groupby('cut')['x'].std().plot()
ddd
```
